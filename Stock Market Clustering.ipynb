```python
# Define which online source to use
data_source = 'yahoo'

# define the start and end dates
start_date = '2015-01-01' 
end_date = '2017-12-31'

# Use pandas_reader.data.DataReader to load the desired stack data
panel_data = data.DataReader(list(companies_dict.values()), data_source, start_date, end_date)

# Print Axes labels
print(panel_data.axes)
```

```python
### Heirarchial Indexing!!
# Unstack: This is used if you want to convert your index into a col for your dataframe 
# Stack:  This is used if you want to convert your col into a index for your dataframe
# swaplevel: This can be used to change the hierarchy of the indices for series/dataframe objects.
```

```python
# We need to do this!
new_data = panel_data.stack(level=0)
```

```python
# Not required. Extra info.
new_data1 = panel_data.unstack(level=0)
```

```python
# The next crucial step. Swapping the index order of df.
new_data.index = new_data.index.swaplevel()
```

```python
# Find the stock open and close data values
stock_close_df = new_data.loc['Close']# Close values
stock_open_df = new_data.loc['Open']# Open values

print(stock_close_df.shape)
print(stock_open_df.shape)
```

```python
stock_close_df.iloc[1]
```

```python
stock_close_df.head()
```

```python
# Calcualate the daily stock movement
stock_close = np.array(stock_close_df).T
stock_open = np.array(stock_open_df).T
```

```python
stock_close.shape
```

```python
row, col = stock_close.shape
```

```python
row, col
```

```python
movements = np.zeros([row, col])
for i in range(row):
    movements[i, :] = np.subtract(stock_close[i,:], stock_open[i,:])
```

```python
movements.shape
```

```python
# Checking the output values
print(stock_close[24][0] )
print("open", stock_open[24][0])
print("diff", movements[24][0])
```

```python
# Visualization - Plot stock movements
plt.clf() # Gives us a blannk starting point
plt.figure(figsize=(18,18))
ax1 = plt.subplot(221)
plt.plot(movements[1][:])
plt.title(stock_close_df.columns[1])

ax2 = plt.subplot(222, sharey=ax1)
plt.plot(movements[0][:])
plt.title(stock_close_df.columns[0])

plt.show()
```

```python
# Import Normlizer
from sklearn.preprocessing import Normalizer

# Create the nomalizer
normalizer = Normalizer()
new = normalizer.fit_transform(movements)

print(movements.mean(), movements.min(), movements.max())
print(new.mean(), new.min(), new.max())
```

```python
# Visualization - Plot stock movements
plt.clf()
plt.figure(figsize=(18,18))
ax1 = plt.subplot(221)
plt.plot(new[1][:])
plt.title(stock_close_df.columns[1])

ax2 = plt.subplot(222,sharey=ax1)
plt.plot(new[0][:])
plt.title(stock_close_df.columns[0])

ax3 = plt.subplot(223)
plt.plot(new[2][:])
plt.title(stock_close_df.columns[2])

plt.show()
```

```python
# Create a pipeline for the whole training process
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans

# Define the normalizer
normalizer = Normalizer()

# Create a Kmeans model - 10 clusters
kmeans = KMeans(n_clusters=8, max_iter=10000)

# Make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer, kmeans)
```

```python
# Fit the pipeline to the daily stock movements
pipeline.fit(movements)
```

```python
# The lesser the intertia the better the model!
# A score of how close each cluster is.
print(kmeans.inertia_)
```

```python
# Predict the cluster labels
labels = pipeline.predict(movements)

# Create a dataframe aligning labels and companies
df = pd.DataFrame({'labels': labels, 'companies': list(stock_close_df.columns)})
```

```python
print(df.sort_values('labels'))
```

```python
# Principal Component Analysis
# Linear dim reduction using singular value decomp of the data.

from sklearn.decomposition import PCA

# Visualize the results on PCA-reduced data
reduced_data = PCA(n_components = 2).fit_transform(new)
print("New shape: ", reduced_data.shape)

# run kmeans on the reduced data
kmeans = KMeans(n_clusters=10)
kmeans.fit(reduced_data)
labels = kmeans.predict(reduced_data)
print("labels shape: ", labels.shape)
print(list(stock_close_df.columns))
# Create a dataframe aligning lables and company names
df = pd.DataFrame({'labels':labels, 'company': list(stock_close_df.columns) })
```

```python
print(df.sort_values('labels'))
```

```python
# Define step size of the mesh
h = 0.01

# Plot the decision boundary
x_min, x_max = reduced_data[:, 0].min()-1, reduced_data[:,0].max() +1
y_min, y_max = reduced_data[:, 1].min()-1, reduced_data[:,1].max()+1

xx,yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
print(xx.shape, yy.shape)
print("inp shape: ", np.c_[xx.ravel(), yy.ravel()].shape )
# Obtain labels for each point in the mesh using our trained model
z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
```

```python
print(reduced_data[:,1].shape)
```

```python
# Define the colomap
cmap = plt.cm.Paired

# Plotting the figure
plt.clf()
plt.figure(figsize=(20,20))
plt.imshow(z, interpolation='nearest',
          extent=(xx.min(), xx.max(), yy.min(), yy.max()),
          cmap=cmap, aspect='auto', origin='lower')
plt.plot(reduced_data[:,0],reduced_data[:,1],'k.',markersize=5)

# Plot the cdntroid of each cluster as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1],marker='x',s=1000, linewidths=3,
           color='w')

plt.title("K-Means clusterig on Stack Market Clustering (PCA-Reduced Data)")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
